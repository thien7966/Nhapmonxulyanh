### Câu 1 (2 điểm): Cho một ảnh bất kỳ (tên ảnh do sinh viên tự đặt, ví dụ: `my_image.jpg`) và thực hiện các yêu cầu sau:

* Viết chương trình sử dụng Bilateral filter để làm mịn ảnh. (0.5 điểm)  
* Viết chương trình sử dụng Canny Edge Detection để xác định biên của hình ảnh. (0.5 điểm)  
* Đổi màu ảnh bằng cách hoán đổi kênh màu theo thứ tự (ví dụ: BGR → BRG) và lưu thành tên dạng `[ten_anh]_swapped.jpg`. (0.5 điểm)  
* Chuyển ảnh sang không gian màu YCrCb và tách riêng 3 kênh Y, Cr, Cb, lưu thành ảnh grayscale tương ứng (`[ten_anh]_Y.jpg`, `[ten_anh]_Cr.jpg`, `[ten_anh]_Cb.jpg`). (0.5 điểm)


```python
#câu 1.1
import cv2

# Bước 1: Đọc ảnh gốc (giả sử ảnh tên 'trai_cay.png')
image = cv2.imread("trai-cay.jpg")

# Bước 2: Áp dụng bộ lọc Bilateral để làm mịn nhưng vẫn giữ viền cạnh rõ
bilateral_filtered = cv2.bilateralFilter(
    image,
    d=9,             # Đường kính pixel vùng lân cận
    sigmaColor=75,   # Độ lệch chuẩn cho không gian màu
    sigmaSpace=75    # Độ lệch chuẩn cho không gian tọa độ
)

# Bước 3: Lưu kết quả ra file
cv2.imwrite("trai_cay_bai1_bilateral.jpg", bilateral_filtered)

# Thông báo hoàn tất
print("Đã lưu ảnh làm mịn tại 'trai_cay_bai1_bilateral.jpg'")






#câu 1.2
import cv2

# Bước 1: Đọc ảnh gốc (đặt đúng tên file ảnh hoặc thay đường dẫn nếu cần)
image = cv2.imread("trai-cay.jpg")

# Kiểm tra ảnh đã đọc thành công chưa
if image is None:
    print("Không tìm thấy ảnh! Vui lòng kiểm tra tên file hoặc đường dẫn.")
    exit()

# Bước 2: Chuyển ảnh sang grayscale (ảnh xám)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Bước 3: Áp dụng thuật toán Canny để phát hiện biên
edges = cv2.Canny(
    gray_image,
    threshold1=100,  # Ngưỡng dưới
    threshold2=200   # Ngưỡng trên
)

# Bước 4: Lưu kết quả ảnh chứa biên
cv2.imwrite("trai_cay_bai2_canny_edges.jpg", edges)

# Thông báo hoàn tất
print("Đã lưu ảnh biên cạnh tại 'trai_cay_bai2_canny_edges.jpg'")




#câu 1.3
import cv2

# Bước 1: Đọc ảnh gốc (đảm bảo ảnh tên 'trai_cay.png' có trong thư mục)
image = cv2.imread("trai-cay.jpg")

# Bước 2: Hoán đổi kênh màu: đổi kênh G và R (BGR → BRG)
swapped_image = image.copy()
swapped_image[:, :, [1, 2]] = swapped_image[:, :, [2, 1]]  # G ↔ R

# Bước 3: Lưu ảnh sau khi hoán đổi kênh
cv2.imwrite("trai_cay_bai3_swapped.jpg", swapped_image)

# Thông báo hoàn tất
print("Đã lưu ảnh với kênh màu hoán đổi tại 'trai_cay_bai3_swapped.jpg'")






#câu 1.4
import cv2

# Bước 1: Đọc ảnh gốc
image = cv2.imread("trai-cay.jpg")

# Kiểm tra nếu ảnh không đọc được
if image is None:
    print("Không tìm thấy ảnh! Hãy kiểm tra tên file hoặc đường dẫn.")
    exit()

# Bước 2: Chuyển đổi ảnh sang không gian màu YCrCb
ycrcb_image = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)

# Bước 3: Tách các kênh Y, Cr, Cb
Y, Cr, Cb = cv2.split(ycrcb_image)

# Bước 4: Lưu các ảnh grayscale tương ứng của từng kênh
cv2.imwrite("trai_cay_bai4_Y.jpg", Y)
cv2.imwrite("trai_cay_bai4_Cr.jpg", Cr)
cv2.imwrite("trai_cay_bai4_Cb.jpg", Cb)

# Thông báo hoàn tất
print("Đã lưu các kênh Y, Cr, Cb thành 3 ảnh:")
print(" - trai_cay_bai4_Y.jpg")
print(" - trai_cay_bai4_Cr.jpg")
print(" - trai_cay_bai4_Cb.jpg")
```

### Câu 2 (4 điểm) Viết một chương trình Python sử dụng OpenCV để tạo menu tương tác cho phép người dùng chọn các kỹ thuật biến đổi hình học và xử lý ảnh nâng cao từ một danh sách, áp dụng đồng thời cho nhiều ảnh.

### Yêu cầu:

1. Menu gồm:  
* Phóng to ảnh (Zoom bằng resize) (0.5 điểm)  
* Xoay ảnh (góc ngẫu nhiên từ 0–360 độ) (0.5 điểm)  
* Lật ảnh ngang (0.5 điểm)  
* Lật ảnh dọc (0.5 điểm)  
* Cắt ảnh (crop ngẫu nhiên vùng giữa ảnh) (0.5 điểm)  
* Thêm viền (padding màu ngẫu nhiên) (0.5 điểm)

2. Chương trình xử lý đồng thời 3 ảnh bất kỳ do sinh viên tự chọn (có thể chọn bằng đường dẫn file hoặc nhập tên ảnh tùy ý). (0.5 điểm)

3. Phím tương ứng để kích hoạt các phương pháp xử lý:  
* Z: Zoom  
* T: Rotate  
* H: Horizontal Flip  
* V: Vertical Flip  
* C: Crop  
* P: Padding (0.5 điểm)

4. Lưu file kết quả với định dạng: `result_[phương pháp]_[tên ảnh gốc].jpg`  
   Ví dụ: `result_crop_cat.jpg`, `result_rotate_image1.jpg` (0.5 điểm)


```python
import cv2
import numpy as np
import random
import os

# === DANH SÁCH 3 ẢNH ĐƯỢC CHỌN ===
image_paths = [
    "trai-cay1.jpg",
    "trai-cay2.jpg",
    "trai-cay3.jpg"
]

# Đọc ảnh vào bộ nhớ
images = [cv2.imread(path) for path in image_paths]
image_names = [os.path.basename(path).split('.')[0] for path in image_paths]

# === HÀM ÁP DỤNG XỬ LÝ ===
def apply_transformation(key, img, name):
    h, w = img.shape[:2]

    if key == ord('z') or key == ord('Z'):
        result = cv2.resize(img, None, fx=2.0, fy=2.0, interpolation=cv2.INTER_LINEAR)
        method = "zoom"

    elif key == ord('t') or key == ord('T'):
        angle = random.randint(0, 360)
        M = cv2.getRotationMatrix2D((w // 2, h // 2), angle, 1)
        result = cv2.warpAffine(img, M, (w, h))
        method = "rotate"

    elif key == ord('h') or key == ord('H'):
        result = cv2.flip(img, 1)
        method = "hflip"

    elif key == ord('v') or key == ord('V'):
        result = cv2.flip(img, 0)
        method = "vflip"

    elif key == ord('c') or key == ord('C'):
        crop_ratio = random.uniform(0.5, 0.7)
        new_h = int(h * crop_ratio)
        new_w = int(w * crop_ratio)
        start_y = (h - new_h) // 2
        start_x = (w - new_w) // 2
        result = img[start_y:start_y + new_h, start_x:start_x + new_w]
        method = "crop"

    elif key == ord('p') or key == ord('P'):
        pad_size = 30
        color = [random.randint(0, 255) for _ in range(3)]
        result = cv2.copyMakeBorder(img, pad_size, pad_size, pad_size, pad_size,
                                    borderType=cv2.BORDER_CONSTANT, value=color)
        method = "padding"

    else:
        return None, None

    output_path = f"result_{method}_{name}.jpg"
    cv2.imwrite(output_path, result)
    return result, output_path

# === GIAO DIỆN MENU TƯƠNG TÁC ===
print("=== MENU ===")
print("Z: Zoom | T: Rotate | H: Horizontal Flip | V: Vertical Flip | C: Crop | P: Padding")
print("ESC để thoát")

while True:
    cv2.imshow("Ảnh 1", images[0])
    key = cv2.waitKey(0)

    if key == 27:  # ESC
        break

    for i in range(3):
        img = images[i]
        name = image_names[i]
        result, path = apply_transformation(key, img, name)
        if result is not None:
            cv2.imshow(f"Kết quả ảnh {i+1}", result)
            print(f" Đã lưu: {path}")
        else:
            print(" Phím không hợp lệ!")

cv2.destroyAllWindows()
```

### Câu 3 (4 điểm) Viết một chương trình Python để xử lý 3 ảnh bất kỳ do sinh viên tự chọn.

* Thêm viền đen 20 pixel cho ảnh đầu tiên. (0.5 điểm)  
* Xoay ảnh thứ hai 45 độ và phóng to 1.5 lần. (0.5 điểm)  
* Tăng kích thước ảnh thứ ba lên 4 lần, sau đó áp dụng Bilateral Filter với tham số tùy chọn. (1.5 điểm)  
* Thay đổi độ sáng và độ tương phản ảnh thứ ba theo công thức:

$$
I_{out}(x, y) = \alpha \cdot I_{in}(x, y) + \beta
$$

Trong đó:  

$$
\alpha \in [0.6, 2.0], \quad \beta \in [-60, 60]
$$

Giá trị đầu ra cần được giới hạn trong khoảng [0, 255] bằng công thức:

$$
I_{out}(x, y) = \text{clip}(I_{out}(x, y), 0, 255)
$$


```python
import cv2
import numpy as np

# Đọc 3 ảnh
img1 = cv2.imread('trai-cay.jpg')
img2 = cv2.imread('dong-vat.jpg')
img3 = cv2.imread('ca.jpg')

# Kiểm tra ảnh có đọc được không
if img1 is None or img2 is None or img3 is None:
    print("Lỗi: Không thể đọc một trong các ảnh.")
    exit()

### 1. Thêm viền đen 20 pixel cho ảnh đầu tiên
img1_border = cv2.copyMakeBorder(
    img1,
    top=20, bottom=20, left=20, right=20,
    borderType=cv2.BORDER_CONSTANT,
    value=[0, 0, 0]  # màu đen
)
cv2.imwrite('result_border_trai_cay.jpg', img1_border)

### 2. Xoay ảnh thứ hai 45 độ và phóng to 1.5 lần
(h2, w2) = img2.shape[:2]
center = (w2 // 2, h2 // 2)
matrix = cv2.getRotationMatrix2D(center, 45, 1.5)  # góc 45 độ, scale 1.5
img2_rotated = cv2.warpAffine(img2, matrix, (int(w2 * 2), int(h2 * 2)))  # khung ảnh lớn hơn để chứa kết quả
cv2.imwrite('result_rotate_dong_vat.jpg', img2_rotated)

### 3. Tăng kích thước ảnh thứ ba lên 4 lần và lọc Bilateral
img3_scaled = cv2.resize(img3, None, fx=4.0, fy=4.0, interpolation=cv2.INTER_LINEAR)
img3_bilateral = cv2.bilateralFilter(img3_scaled, d=9, sigmaColor=75, sigmaSpace=75)
cv2.imwrite('result_bilateral_ca.jpg', img3_bilateral)

### 4. Thay đổi độ sáng và tương phản ảnh thứ ba theo công thức
alpha = 1.8   # độ tương phản trong khoảng [0.6, 2.0]
beta = 30     # độ sáng trong khoảng [-60, 60]
img3_modified = cv2.convertScaleAbs(img3_bilateral, alpha=alpha, beta=beta)
cv2.imwrite('result_brightness_ca.jpg', img3_modified)

print("Xử lý hoàn tất! Các ảnh đã được lưu với tên bắt đầu bằng 'result_'.")
```
